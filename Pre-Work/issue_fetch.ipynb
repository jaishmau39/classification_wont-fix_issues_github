ISSUE DETAILS FOR ALL USERS AND REPOSITORIES

```python
#!/usr/bin/python
from github import Github

from perceval.backends.core.github import GitHub
 
# Set this to a personal access token.
# - Select "repo" for the oauth scopes.
# See https://help.github.com/articles/creating-a-personal-access-token-for-the-command-line/
g = Github('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')
 
repositories = set()
 
# Note: Gets rate limited and fails if too many hits
content_files = g.search_code(query='https://api.github.com/search/repositories?q=stars:>=10000+language:python&sort=stars&order=desc')
#content_files = g.search_code(query='https://github.com/search?q=language%3AC%23&type=Repositories&ref=advsearch&l=&l=C%23')
for content in content_files:
    repositories.add(content.repository.name)
    repositories.add(content.repository.owner.login)
    #repositories.add(content.repository.full_name)
    

    github_4 = GitHub(content.repository.owner.login,content.repository.name, api_token=['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'])

    for issue in github_4.fetch():
        print(issue['data']['labels'])


```
